```python
import random
import time

# Define the cards
card_deck = [
    {"Name": "Diablo", "Health": 100, "Attack": 90, "Defense": 60},
    {"Name": "Medusa", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Jester", "Health": 120, "Attack": 60, "Defense": 90},
    {"Name": "Troll", "Health": 150, "Attack": 40, "Defense": 94},
    {"Name": "Specter", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Mist", "Health": 100, "Attack": 75, "Defense": 65},
    {"Name": "Savage", "Health": 100, "Attack": 90, "Defense": 50},
    {"Name": "Marauder", "Health": 100, "Attack": 85, "Defense": 50},
    {"Name": "Wimp", "Health": 110, "Attack": 40, "Defense": 85},
    {"Name": "Sorcerer", "Health": 100, "Attack": 70, "Defense": 55}
]

deck_size = 5

# Initialize decks
player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample(card_deck, deck_size)

def print_card(card):
    print(f"Name: {card['Name']}")
    print(f"Health: {card['Health']}")
    print(f"Attack: {card['Attack']}")
    print(f"Defense: {card['Defense']}")

def fight(player_deck, opponent_deck):
    while player_deck and opponent_deck:
        print("\nPlayer's Deck:")
        for index, card in enumerate(player_deck):
            print(f"{index + 1}: {card['Name']}")

        opponent_card = random.choice(opponent_deck)
        print("\nOpponent's Card:")
        print_card(opponent_card)

        chosen_index = int(input("Choose a card to fight (by index, 1-5): ")) - 1
        if chosen_index < 0 or chosen_index >= len(player_deck):
            print("Invalid choice. Please choose a valid card.")
            continue

        player_card = player_deck[chosen_index]

        print(f"\n{player_card['Name']} attacks {opponent_card['Name']}.")

        # Calculate damage
        damage = player_card['Attack'] - opponent_card['Defense']
        damage = max(0, damage)  # Ensure no negative damage
        opponent_card['Health'] -= damage
        opponent_card['Defense'] = max(0, opponent_card['Defense'] - player_card['Attack'])

        if opponent_card['Health'] <= 0:
            print(f"{opponent_card['Name']} is defeated!")
            opponent_deck.remove(opponent_card)
        else:
            print(f"{opponent_card['Name']} now has {opponent_card['Health']} Health and {opponent_card['Defense']} Defense.")

        time.sleep(3)  # Pause for effect

        if not opponent_deck:
            print("\nPlayer wins!")
            break

        print("\nOpponent's turn.")
        opponent_card = random.choice(opponent_deck)
        print(f"{opponent_card['Name']} attacks {player_card['Name']}.")

        # Opponent's attack logic
        damage = opponent_card['Attack'] - player_card['Defense']
        damage = max(0, damage)
        player_card['Health'] -= damage
        player_card['Defense'] = max(0, player_card['Defense'] - opponent_card['Attack'])

        if player_card['Health'] <= 0:
            print(f"{player_card['Name']} is defeated!")
            player_deck.remove(player_card)
        else:
            print(f"{player_card['Name']} now has {player_card['Health']} Health and {player_card['Defense']} Defense.")

        time.sleep(3)  # Pause for effect

        if not player_deck:
            print("\nOpponent wins!")
            break

        print("\nNew round in 3 seconds...")
        time.sleep(3)

if __name__ == "__main__":
    fight(player_deck, opponent_deck)
```
